## Mobike数据集的简单评估和清理

```python
#载入所需数据包
import pandas as pd
import numpy as np
```

```python
#读入数据集
df = pd.read_csv('mobike_shanghai_sample_updated.csv')
```

```python
#查看数据集信息
df.info()
```

```python
#设置显示每列里的所有内容，而不会被默认砍掉
pd.set_option('display.max_colwidth', -1)

#查看前10行数据
df.head(10)
```

```python
#检查数据集是否有重复订单
df.orderid.duplicated().sum()
```

```python
#查看总用户数
df.userid.nunique()
```

```python
#查看总单车数
df.bikeid.nunique()
```

以上可见，数据集涉及到102361条订单数据，16887个用户和79062辆单车。

由于骑行轨迹列，对我提出的5个问题的探究没有太多帮助，以下删除该列。

```python
#删除骑行轨迹列
df = df.drop(['track'],axis=1)
```

```python
#将start_time和end_time列转成正确的时间格式，以便后续进行计算
df.start_time = pd.to_datetime(df.start_time)
df.end_time = pd.to_datetime(df.end_time)
```

```python
#计算骑行时长，存入新列duration中
df['duration'] = df.end_time - df.start_time
df['duration']
```

```python
#将duration列的时间格式转化成分钟数
df['duration'] = df['duration'].apply(lambda x: x/np.timedelta64(1,'m'))
```

```python
#根据start_time排序
df.sort_values('start_time')
```

以上可以看出数据集记录了2016年8月1日到8月31日的订单，即8月整月的数据。

```python
#查看用户骑行次数
df.userid.value_counts()
```

```python
#对骑行时长进行排序
df.duration.sort_values(ascending=False)
```

```python
#查看骑行时长最长的数据
df.iloc[61444]
```

以上数据可见，起始坐标基本没变，骑行时间确持续了3天多。明显为没有锁车导致的。

```python
#查看另外一个骑行时长较长的行
df.iloc[36431]
```

以上行也是同样的情况，应该是没有锁车的异常情况。

此数据集中有不少类似的异常情况，因此在可视化的过程中，需要考虑到针对此类数据的统计。
